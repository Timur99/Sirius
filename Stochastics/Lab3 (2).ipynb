```python
from scipy.stats import norm
import matplotlib.pyplot as plt
from math import sqrt, exp
import random
import numpy as np
import seaborn as sns



def expected_OU(X_0, theta1, theta2, theta3, t):
    expected_value = (theta1 / theta2) + (X_0 - theta1 / theta2) * np.exp(-theta2 * t)
    return expected_value

def variance_OU(X_0, theta1, theta2, theta3, t):
    variance = ((theta3 ** 2) / (2 * theta2)) * (1 - np.exp(-2 * theta2 * t))
    return variance

theta1 = 2
theta2 = 0.2
theta3 = 0.15
N = 1000
dW = np.random.normal(0, np.sqrt(1 / N), N)
X_ou = np.zeros(N)
X_ou[0] = 10
Y_ou = np.zeros(N)
Y_ou[0] = X_ou[0]

Yt = np.zeros(N)
Yt[0] = X_ou[0]


for i in range(1, N):

    X_ou[i] = X_ou[i - 1] + (theta1 - theta2 * X_ou[i - 1]) * i/N + theta3 * dW[i]

    ito_sum =[ (exp ( - theta2 *(1/N ))) *dW[i] for i in range (1,N ) ]


    Yt[i] = (theta1 / theta2) + (Yt[i-1] - theta1 / theta2) * np.exp(-theta2 * i/N) + theta3 * sum(ito_sum [0: i-1 ])

#ito_sum =[ exp ( - theta *( t [ i ] - t [i -1]) ) *( W [ i ] - W [i -1]) for i in range (1,N ) ]
#Y[i] = [ X [0]* exp ( - theta * t [ i ]) + sigma* sum(ito_sum [0: i ]) for i in range ( N ) ]
#X[i] = [ X [0]]+ Y[i]


    loc = expected_OU(X_ou[i], theta1, theta2, theta3, i / N)
    scale = variance_OU(X_ou[i], theta1, theta2, theta3, i / N)
    Y_ou[i] = np.random.normal(loc, scale, 1)[0]

# 3 задание
fig, ax = plt.subplots()
ax.plot()
#ax.plot(np.linspace(0, N, N), Yt, 'c', linewidth=2, label='random variable')
ax.plot(np.linspace(0, N, N), X_ou, 'b', linewidth=2, label='process OU')
plt.show()


# 4 и 5 задания


fig, ax = plt.subplots()
ax.plot()
ax.plot(np.linspace(0, N, N), X_ou, 'b', linewidth=2, label='process OU')
ax.plot(np.linspace(0, N, N), Y_ou, 'c', linewidth=2, label='random variable expected and variance')
plt.show()




fig, ax = plt.subplots()
ax.plot(np.sort(X_ou), [i / N for i in range(N)], 'b', linewidth=2, label='distribution , process OU')
ax.plot(np.sort(Y_ou), [i / N for i in range(N)], 'c', linewidth=2, label='distribution , random variable')
plt.show()




sns.kdeplot (X_ou)
sns.kdeplot (Y_ou)
```
