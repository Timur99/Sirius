  # $$W(t_{i}) = W(t_{i-1}) + z \sqrt{Δ{t}}$$ ,




```python

import random
import numpy as np
import math
import matplotlib.pyplot as plt

N = 100
T = 1
X =[i /N for i in range (N+1)]

def winir () :
    W =[]
    W +=[0]
    Delta = T / N
    N_n = np.random.normal (0, 1, N+1)
    for i in range (1,N+1) :
        W +=[ W [i -1]+ N_n[i] * math.sqrt ( Delta ) ]   #  A simulated path of the Wiener process
    return W


W = winir()
plt.figure ( figsize =(10, 7) )
plt.grid ()
plt.plot (X , W , linewidth =2.0)
plt.title ("Wiener process")
plt.xlabel ("t")
plt.ylabel ("W")
plt.show ()
```

  # $$W(t) = W(t, \omega) = \ \sum_{i=0}^n Z_{i}(\omega) \phi_{i}(t)$$ ,
  # $$\phi_{i}(t) = {(2 \sqrt{2T}) \over ((2i + 1) \pi)} \sin({(2i + 1) \pi t \over 2T})$$"

```python
# Приближение Кархунена-Леве к траектории винеровского процесса
import numpy as np

import matplotlib.pyplot as plt
import math
import random
from math import sqrt, sin, pi

def phi (i ,t , T ) :
    return (2* sqrt (2* T ) ) /(( 2* i +1) * pi ) * sin ((( 2* i +1) * pi * t ) /(2* T ) )

def sum_W (t , T, Z ) :
    Sum =0
    for i in range (len( Z ) ) :
        Sum += Z [ i ]* phi (i ,t , T )
    return Sum


T =1
N =100
t = [ i /N for i in range (N+1) ]
nn = [10,50,100]
n = 10


def WWW(n):
    W =[]
    Z = np . random . normal (0, 1, n )
    for i in range ( N +1) :
        W +=[ sum_W ( t [ i ] , T, Z ) ]
    return W

fig, ax = plt.subplots()

for j in nn:
  ax.plot(t, WWW(j), linewidth=2.0)
  ax.grid(linewidth = 1)
  #plt.legend()
  fig.set_figwidth(10)
  fig.set_figheight(7)
  plt.title ('Wiener process ')
  plt.xlabel ("t")
  plt.ylabel ("W")
  plt.legend([10,50,100], loc ='upper right')
plt.show()
```

```python
import numpy as np

import matplotlib.pyplot as plt
from math import sqrt, exp
import random


r =1
sigma =0.5
x =10
N =100
T =1
Delta = T / N
W =[0]
t = np . linspace (0,T , N +1)  #сетка времени

#for i in range ( N ) : #винеровский процесс
#    W +=[ W [i -1]+ np . random . normal (0, 1, 1) * sqrt ( Delta ) ]
 

W = np.zeros(N+1)
for i in range(N+1):  # генерим винеровский процесс
    W[i] +=[W[i-1] + np.random.normal(0, 1, 1) * sqrt ( Delta )]


S = []
for i in range (len(W)):
    S.append(x * exp (( r - sigma **2/2) * t [ i ] + sigma * W [ i ])) #for i in range (len( W ) ) ] # геометрическое броуновское движение

plt.figure ( figsize =(10, 7) )
plt.grid ()
plt.plot (t , S , linewidth =2.0)

plt.title ("geometric ␣ Brownian ␣ motion ")
plt.xlabel ("t")
plt.ylabel ("S")
plt.show ()
```

```python
import numpy as np

import matplotlib.pyplot as plt
from math import sqrt, exp
import random



N =100
T =1
Delta = T / N
W = [0]
t = list (np.linspace (0,T , N) ) #сетка времени

x = 0
y = -1

W = np.zeros(N)
for i in range(N):  # генерим винеровский процесс
    W[i] +=[W[i-1] + np.random.normal(0, 1, 1) * sqrt ( Delta )]
  
print(W)

def BBB(x,y):
    BB = []
    for i in range(len(W)):
        BB.append(x + W[i] - t[i] / T * (W[N-1] - y + x))     #  Brownian bridge
    return BB

plt.figure ( figsize =(10, 7) )
plt.grid ()
plt.plot (t , BBB(x,y) , linewidth =2.0)
plt.scatter ( t [0] , x , color ="red",s =50)
plt.scatter ( t [len( t ) -1] , y , color ="red", s =50)
plt.title ("Brownian bridge")
plt.xlabel ("t")
plt.ylabel ("BB")
plt.show ()

    
```
