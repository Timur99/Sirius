```python
import numpy as np

import matplotlib.pyplot as plt
from math import sqrt, exp
import random
N =100
theta1 = 2.00
theta2 = 0.20
theta3 = 0.15
theta4 = 0.5

def BM () :
  W =[]
  T =1
  W +=[0]
  Delta = T / N
  X =[ i /N for i in range (N+1) ]
  for i in range (1,len( X ) ) :
    W +=[ W [i -1]+ np . random . normal (0, 1, 1) * sqrt ( Delta ) ]
  return W

W = BM ()

t =[ i /N for i in range (N+1) ]

x =100

dt = 1/N
X = []
for w in W:
  X.append((theta1 + theta2*x)*dt + theta3 * (x**theta4) * w)



plt . figure ( figsize =(10, 7) )
plt . grid ()
plt . plot (t , X )
plt . title ("CKLS  process")
plt . xlabel (" Time ")
plt . ylabel ("X")
plt . show ()
```

For the Ornstein-Uhlenbeck process
dXt = θ(µ − Xt)dt + σdWt, θ = 1, µ = 1.2, σ = 0.3
generate three paths: $X_0 = 0.0, 2.0, N\left(µ, \frac{σ^2}{2θ}\right)$
and plot in graph. Define
a path with the stationary distribution from the graph.

```python
import numpy as np

import matplotlib.pyplot as plt
from math import sqrt, exp
import random
N =100
theta1 = 1
r = 1.2
sigma = 0.3


n = np.random.normal(r, (sigma ** 2) / 2 * theta1)
print(n)

X0 = [0.0,2.0, n]

def BM () :
  W =[]
  T =1
  W +=[0]
  Delta = T / N
  X =[ i /N for i in range (N+1) ]
  for i in range (1,len( X ) ) :
    W +=[ W [i -1]+ np . random . normal (0, 1, 1) * sqrt ( Delta ) ]
  return W

W = BM ()


#t =[ i /N for i in range (N) ]
t = np.linspace(0, N, N)

fig, ax = plt.subplots()

for j in X0:
  delta = 1/N
  X = np.zeros(N)
  X[0] = j
  for i in range(1,len(X)):
    X[i] = X[i-1] + (theta1 * (r - X[i-1])*dt + sigma * np.random.normal(0, np.sqrt(delta)))
  ax.plot(t, X, label=round(j, 2), linewidth=2.0)
  ax.grid(linewidth = 1)
  plt.legend()
  fig.set_figwidth(10)
  fig.set_figheight(7)
  plt.title('График по сигме')
  plt.xlabel('Цена актива')
  plt.ylabel('Цена опциона по t')
plt.show()




  

```
